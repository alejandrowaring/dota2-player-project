```python
import pymongo
import json
```

```python
conn = 'mongodb://localhost:27017'
client = pymongo.MongoClient(conn)
dotadb = client.heroesDB
```

```python
import json
```

```python
with open("./heroes.json") as json_heros_file:
    data = json.load(json_heros_file)
    for i in data:
        dotadb.heroes.insert_one(i)
```

```python
heroes = dotadb.heroes.find()
for i in heroes:
    print(i['localized_name'])
```

```python
with open("./items.json") as json_items_file:
    data = json.load(json_items_file)
    # print(data)
    for i in data:
        dotadb.items.insert_one(data[i])
```

```python
items = dotadb.items.find()
for i in items:
    print(i['dname'])
```
